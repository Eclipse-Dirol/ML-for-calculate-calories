```python
import pandas as pd
import sqlite3 as sql
```

```python
data = pd.read_csv('data/train.csv')
data_test = pd.read_csv('data/test.csv')
display(data.shape)
display(data_test.shape)
```

```python
full_data = pd.concat([data, data_test], ignore_index=True)
```

```python
display(full_data.shape)
display(full_data.isna().sum())
```

```python
def create_db(data: pd.DataFrame):
    data_for_db = data.copy()
    
    conn = sql.connect('database.db')
    cursor = conn.cursor()
    
    cursor.execute("""
        CREATE TABLE data (
            id INTEGER PRIMARY KEY,
            Sex TEXT,
            Age INTEGER,
            Height REAL,
            Weight REAL,
            Duration REAL,
            Heart_Rate REAL,
            Body_Temp REAL,
            Calories REAL
        )
""")
    conn.commit()
    conn.close()
```

```python
create_db(data = full_data)
```
