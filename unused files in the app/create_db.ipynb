```python
import sqlite3 as sql
```

```python
BASE_DIR = "P:/pet ML"
```

```python
def create_db():
    conn = sql.connect(f'{BASE_DIR}/data/database.db')
    cursor = conn.cursor()
    
    cursor.execute("""
        CREATE TABLE data (
            id INTEGER PRIMARY KEY,
            Sex_male INTEGER,
            Sex_female INTEGER,
            Age INTEGER,
            Height REAL,
            Weight REAL,
            Duration REAL,
            Calories REAL
        )
""")
    conn.commit()
    conn.close()
```

```python
create_db()
```
